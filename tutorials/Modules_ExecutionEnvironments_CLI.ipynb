<h3>Actually computing the daily-averages...</h3>
<h4>Looping through each month and day, to calculate the daily-average temperature</h4>

<p>
<blockquote>
<ul>
<li>Recall that each month is represented by an <strong><code>integer</code></strong> in the new <strong><code>DataFrame</code></strong></li>
<li>...and the months are represented by month <strong><code>string</code></strong>s in the newly constructed data-container</li>

````python
for each_month in pd.unique(hobo2['Month']):
    if each_month == 1:
        for each_day in pd.unique(hobo2['Day']\
                                  [hobo2['Month'] == each_month]):
            avgTT_by_month['January'].append\
            (np.nanmean(hobo2['Temperature'][hobo2['Month'] == each_month][hobo2['Day'] == each_day]))
    elif each_month == 2:
        for each_day in pd.unique(hobo2['Day']\
                                  [hobo2['Month'] == each_month]):
            avgTT_by_month['February'].append\
            (np.nanmean(hobo2['Temperature'][hobo2['Month'] == each_month][hobo2['Day'] == each_day]))
    elif each_month == 3:
        for each_day in pd.unique(hobo2['Day']\
                                  [hobo2['Month'] == each_month]):
            avgTT_by_month['March'].append\
            (np.nanmean(hobo2['Temperature'][hobo2['Month'] == each_month][hobo2['Day'] == each_day]))
    elif each_month == 4:
        for each_day in pd.unique(hobo2['Day']\
                                  [hobo2['Month'] == each_month]):
            avgTT_by_month['April'].append\
            (np.nanmean(hobo2['Temperature'][hobo2['Month'] == each_month][hobo2['Day'] == each_day]))
    elif each_month == 5:
        for each_day in pd.unique(hobo2['Day']\
                                  [hobo2['Month'] == each_month]):
            avgTT_by_month['May'].append\
            (np.nanmean(hobo2['Temperature'][hobo2['Month'] == each_month][hobo2['Day'] == each_day]))
    elif each_month == 6:
        for each_day in pd.unique(hobo2['Day']\
                                  [hobo2['Month'] == each_month]):
            avgTT_by_month['June'].append\
            (np.nanmean(hobo2['Temperature'][hobo2['Month'] == each_month][hobo2['Day'] == each_day]))
    elif each_month == 7:
        for each_day in pd.unique(hobo2['Day']\
                                  [hobo2['Month'] == each_month]):
            avgTT_by_month['July'].append\
            (np.nanmean(hobo2['Temperature'][hobo2['Month'] == each_month][hobo2['Day'] == each_day]))
    else:
        for each_day in pd.unique(hobo2['Day']\
                                  [hobo2['Month'] == each_month]):
            avgTT_by_month['August'].append\
            (np.nanmean(hobo2['Temperature'][hobo2['Month'] == each_month][hobo2['Day'] == each_day]))
````

<h3>2. Re-organizing & creating new <code>DataFrame</code></h3>

````python
def convert_data(data):
    '''
    Routine to convert HOBO-Logger data with multiple observations for each day
    of data-logging.
    '''
    import datetime
    import pandas as pd
    
    ##  > Parsing Time Stamps:
    yy, mm, dd = [], [], []
    for ts in pd.to_datetime(data['Date Time, GMT-08:00']):
        yy.append(ts.year)
        mm.append(ts.month)
        dd.append(ts.day)
    t = []  ## empty List
    
    ##  > Creating new datetime vector:
    for year, month, day in zip(yy, mm, dd):
        t.append(datetime.date(year, month, day))
        
    ##  > Organizing new data fields from dictionary
    data_fields = {'DateTime' : t, 
                   'Year' : yy, 
                   'Month' : mm, 
                   'Day' : dd, 
                   'Temp' : data['Temp']}
    
    ##  > Creating new DataFrame and returning it.
    data_reformatted = pd.DataFrame(data_fields)
    return data_reformatted
````

<h3>3. Getting months sampled in reorganized/converted dataset.</h3>

````python
def get_months_sampled(data):
    '''
    Uses the 'convert_data()' function to create new Pandas Data Frame 
    with a 'Month' column (by number).
    
    Depending on what number the month is, 
    '''
    converted_data = convert_data(data)
    months = []
    for each_month_no in pd.unique(converted_data['Month']):
        ##  > Recall months are organized by numbwer-integers
        if each_month_no == 1:
            months.extend(['January'])
        elif each_month_no == 2:
            months.extend(['February'])
        elif each_month_no == 3:
            months.extend(['March'])
        elif each_month_no == 4:
            months.extend(['April'])
        elif each_month_no == 5:
            months.extend(['May'])
        elif each_month_no == 6:
            months.extend(['June'])
        elif each_month_no == 7:
            months.extend(['July'])
        elif each_month_no == 8:
            months.extend(['August'])
        elif each_month_no == 9:
            months.extend(['September'])
        elif each_month_no == 10:
            months.extend(['October'])
        elif each_month_no == 11:
            months.extend(['November'])
        else:
            months.extend(['December'])
    return months
````

<h3>4. Calculating Daily Average Temperature Data</h3>

````python
def calc_daily_avgTT(data):
    '''
    Calculates daily average temperature from HOBO-Logger recording @
    1-hr frequency.
    '''
    import datetime
    import numpy as np
    import pandas as pd
    
    converted_data = convert_data(data)
    months_sampled = get_months_sampled(data)
    avgTT_by_month = {}  ##  empty dictionary
    
    for each_month in months_sampled:
        avgTT_by_month[each_month] = []  ##  empty List TO-contain the daily average Temperatures
       
    for each_month in pd.unique(converted_data['Month']):
        if each_month == 1:
            ##  Loop through each day in each month-sampled, in the converted data.
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['January'].append(np.mean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                         [converted_data['Day'] == each_day]))
        elif each_month == 2:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['February'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 3:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['March'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 4:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['April'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 5:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['May'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 6:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['June'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 7:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['July'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 8:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['August'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 9:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['September'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 10:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['October'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 11:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['November'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
        elif each_month == 12:
            for each_day in pd.unique(converted_data['Day'][converted_data['Month'] == each_month]):
                avgTT_by_month['December'].append(np.nanmean(converted_data['Temp'][converted_data['Month'] == each_month]\
                                                             [converted_data['Day'] == each_day]))
    return avgTT_by_month
````
